```python
import pandas
from sklearn import tree
```

```python
data = pandas.read_csv('titanic.csv')

data = data[pandas.notnull(data['Survived'])]
data = data[pandas.notnull(data['Age'])]
data = data[pandas.notnull(data['Fare'])]
data = data[pandas.notnull(data['Pclass'])]
data = data[pandas.notnull(data['Sex'])]


data['SexN'] = data['Sex'].map({'female': 1, 'male': 0})

my_data = data[['Pclass' , 'Fare', 'Age', 'SexN', 'Survived']]
```

```python
X = my_data[['Pclass' , 'Fare', 'Age', 'SexN']].values
Y = my_data[['Survived']].values.flatten()
clf = tree.DecisionTreeClassifier(random_state=123)

my_data.shape

my_data
```

```python
clf = clf.fit(X,Y)
```

```python
clf.feature_importances_
```

```python
print(dict(zip(my_data.columns, clf.feature_importances_)))
```
